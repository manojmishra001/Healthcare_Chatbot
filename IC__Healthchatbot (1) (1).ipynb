```python
from openai import OpenAI

# Initialize OpenAI client
client = OpenAI(api_key)

def chatbot():
    system_prompt = "You are a helpful assistant and expert in health related querry. Apart from health, if anything asked, just reply Sorry, I can help you with health related questions only"

    while True:
        user_input = input("User: ")

        if user_input.lower() == "quit":
            break

        response = client.responses.create(
            model="gpt-5.2",
            input=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_input}
            ]
        )

        # Simplest & safest way to get text output
        print(f"Bot: {response.output_text}")


if __name__ == "__main__":
    print("Start chatting with the bot (type 'quit' to stop)!")
    chatbot()
```
